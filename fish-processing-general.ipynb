## Preprocessing

```python
#target countries: {spain, usa, china, south Africa, comoros, yemen}
# country and year are selected 
# fish is the label: area predictions are summed together
# this would be a dataframe -> turned into a csv and sent to enne

# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, KBinsDiscretizer
from sklearn.compose import TransformedTargetRegressor
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
initial_df = pd.read_json("captured_marine_fish.json")
initial_df.head()
```

```python
#make area code string for future traintest split
initial_df["AREA_CODE"] = initial_df["AREA_CODE"].astype(str)
```

```python
# have user select a country input("Please Select a Country")
# set to China for testing
selected_country = "Spain"
#input later but for the sake of testing: "Tunas, bonitos, billfishes"
#selected_fish = "Tunas, bonitos, billfishes"
```

```python
user_df = initial_df[initial_df['COUNTRY_NAME'] == selected_country]
#user_df = user_df[user_df["SPECIES_ISSCAAP_GROUP"] == selected_fish]
user_df
```

```python
# drop country name for now; deal with filter later 
# drop scientific name for now; make exact later
processing_df = user_df.drop(["COUNTRY_NAME","COUNTRY_GEOGRAPHIC_REGION","COUNTRY_CONTINENT","SPECIES_SCIENTIFIC_NAME","SPECIES_CPC_GROUP","MEASURE_NAME","AREA_NAME","AREA_OCEAN_GROUP","MEASURE_UNIT"], axis=1)
processing_df.head()
# pca 
# feature importance 
```

```python
new_df = processing_df[processing_df['STATUS_NAME'] != 'NEGLIBILE']
next_df = new_df.drop(["STATUS_NAME"], axis=1)
final_df = next_df[next_df['VALUE'] != 0.0]
final_df.head()
```

```python
to_test = pd.get_dummies(final_df)
X = to_test.drop(['VALUE','COUNTRY_ECONOMIC_CLASS_Developed countries or areas','SPECIES_ISSCAAP_GROUP_Marine fishes not identified'], axis=1)
y = to_test['VALUE']
X.columns
```

```python
import numpy as np
#uniform All bins in each feature have identical widths.; ordinal ensures they're returned as floats or ints
trans = KBinsDiscretizer(n_bins=20, encode='ordinal', strategy='uniform')
reshape = y.array.reshape(-1,1)
#reshape = [y].reshape(-1,1)
data = trans.fit_transform(reshape)
dataset = pd.DataFrame(data)
# histograms of the variables
new_y = np.ravel(data)
new_y
```

```python
edges = trans.bin_edges_[0]
bins = []
for i in edges: 
    bins.append(i)
bins
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, new_y, random_state=1)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
training_score = clf.score(X_train_scaled, y_train)
testing_score = clf.score(X_test_scaled, y_test)
```

```python
#organize into  
prediction_years = [2020,2030,2040,2050]
full_text = X.columns
species_list = []
area_list = []

for i in full_text:
    if 'SPECIES' in i:
        holder1 = i.split('_')
        species_list.append(holder1[3])
    if 'AREA' in i:
        holder2 = i.split('_')
        area_list.append(holder2[2])

num_species = len(species_list) 
num_areas = len(area_list)
prediction_fish = [0] * num_species
prediction_ocean = [0] * num_areas
```

```python
prediction_values = {}
for i in prediction_years:
    prediction_values[i] = {}
    
    for j, index in enumerate(prediction_fish):
        key_name = species_list[j]
        prediction_fish[j] = 1
        prediction_values[i][key_name] = {}

        for k, index in enumerate(prediction_ocean):
            next_key_name = area_list[k]
            prediction_ocean[k] = 1
            full_list = [i] + prediction_fish + prediction_ocean
            prediction_values[i][key_name][next_key_name] = clf.predict([full_list])
            #reset
            prediction_ocean[k] = 0

        #reset
        prediction_fish[j] = 0
```

```python
#convert dictionary into something more readable for csv purposes

dict_df = pd.DataFrame.from_dict({(i,j,k): prediction_values[i][j] 
                           for i in prediction_values.keys() 
                           for j in prediction_values[i].keys()},
                       orient='index')
#reset index will convert them into columns
fish_df = dict_df.reset_index()
fish_df
```

```python
fish_df2 = fish_df.rename(columns={'level_0': 'Year', 'level_1': 'Fish Species'})
fish_df2 = fish_df2.drop(['level_2'], axis = 1)
for column in fish_df2:
    if column != 'Year' and column != 'Fish Species':
        fish_df2[column] = fish_df2[column].astype(int)
        
fish_df2.head()
```

```python
labels = []

for i, index in enumerate(bins):
    if i != len(bins) - 1:
        str_holder = str(index) + " to " + str(bins[i+1])
        labels.append(str_holder)
```

```python
area_indcies = np.arange(2, len(fish_df2.columns))
est_sum = 0
new_col = []
for row in fish_df2.iterrows():
    for i in area_indcies:
        est_sum += row[1][i]
        #est_sum += row[1][area_indcies[i]]
    for j, index in enumerate(labels):
        if j == est_sum:
           new_col.append(index)
    #reset
    est_sum = 0

#sums
fish_df2["Estimated Range of Fish Caught"] = new_col
fish_df2.head()
```
